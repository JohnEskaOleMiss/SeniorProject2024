```python
import pandas as pd
import torch
import numpy as np
from transformers import DistilBertTokenizer, DistilBertForSequenceClassification

def load_model_and_predict(email_text):
    #load the trained model and tokenizer
    model = DistilBertForSequenceClassification.from_pretrained('./distilbert-phishing-detection')
    tokenizer = DistilBertTokenizer.from_pretrained('./distilbert-phishing-detection')

    #tokenize the email text
    inputs = tokenizer(email_text, return_tensors="pt", truncation=True, padding=True, max_length=512)

    #predict using the model
    with torch.no_grad():
        model.eval()
        outputs = model(**inputs)
        logits = outputs.logits

    #get the predicted label
    predicted_class = torch.argmax(logits, dim=1).item()

    return predicted_class

def bert():
  #instantiate variables and dataframe
  df = pd.read_csv('test_data_BERT.csv')
  false_positive = 0
  false_negative = 0
  true_positive = 0
  true_negative = 0
  counter = 0

  #loop through and determine the prediction vs actual label
  for i in range(len(df)):
    email_text = df.iloc[i]['Email Text']
    label = df.iloc[i]['Email Type']
    predicted_class = load_model_and_predict(email_text)

    #calculate fp, fn, tp, tn
    if predicted_class == 1 and label == 'Phishing Email':
      true_positive += 1
    elif predicted_class == 0 and label == 'Safe Email':
      true_negative += 1
    elif predicted_class == 1 and label == 'Safe Email':
      false_positive += 1
    elif predicted_class == 0 and label == 'Phishing Email':
      false_negative += 1


    if(counter%100 == 0):
      print(counter)
    counter += 1

  #calculate accuracy
  accuracy = (true_positive + true_negative) / (true_positive + true_negative + false_positive + false_negative)
  print("Accuracy:", accuracy)

  #print confusion matrix
  array = np.array([[true_positive, false_positive], [false_negative, true_negative]])
  print("Confusion Matrix:")
  print(array)
```

```python
def main():
  bert()

main()
```
